```python
import urllib
```

```python
import urllib.request
```

```python
from bs4 import BeautifulSoup
```

```python
def make_soup(url):
    thepage = urllib.request.urlopen(url)
    soupdata = BeautifulSoup(thepage,"html.parser")
    return soupdata
```

```python
soup = make_soup("https://www.topvalue.com/")
print(soup)
```

```python
for img in soup.findAll('div'):
        print(img)
   
    
```
